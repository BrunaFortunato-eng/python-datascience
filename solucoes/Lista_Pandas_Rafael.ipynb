```python
#################### 06
import pandas as pd
import numpy as np

def HowManyCilindros(cilindros):
    if type(cilindros) == str:
        modelos = DataCarro.loc[DataCarro["num-of-cylinders"] == cilindros]
    elif cilindros == 2:
        modelos = DataCarro.loc[DataCarro["num-of-cylinders"] == "two"]
    elif cilindros == 3:
        modelos = DataCarro.loc[DataCarro["num-of-cylinders"] == "three"]
    elif cilindros == 4:
        modelos = DataCarro.loc[DataCarro["num-of-cylinders"] == "four"]
    elif cilindros == 5:
        modelos = DataCarro.loc[DataCarro["num-of-cylinders"] == "five"]
    elif cilindros == 6:
        modelos = DataCarro.loc[DataCarro["num-of-cylinders"] == "six"]
    elif cilindros == 8:
        modelos = DataCarro.loc[DataCarro["num-of-cylinders"] == "eight"]
    elif cilindros == 12:
        modelos = DataCarro.loc[DataCarro["num-of-cylinders"] == "twelve"]
    modelos.to_csv("{}cilindros.csv".format(cilindros),index=False)

HowManyCilindros(2)
```
