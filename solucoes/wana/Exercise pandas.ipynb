```python
# 2 Encontre o nome da empresa de carros mais cara

df3 = pd.read_csv('carros.csv', sep=',')

df3.set_index('index', inplace=True)
maxe = df3['price'].max()
df_emp = df3.loc[df3['price'] >= maxe, 'company'] 

df_emp
```

```python
# 3 mostre a média de preços
df4 = pd.read_csv('carros.csv', sep=',')
df4.set_index('index', inplace=True)
mean_price = df4['price'].mean()
mean_price
```

```python
# 4 mostre o valor mais alto e mais baixo de horsepower
df5 = pd.read_csv('carros.csv', sep=',')
df5.set_index('index', inplace=True)
max_horse = df5['horsepower'].max()
min_horse = df5['horsepower'].min()
print('max:',max_horse, 'min: ',min_horse)
```
